```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#2D Plots
x = -6.283, -5.498, -4.712, -3.927, -3.142, -2.356, -1.571, -.7854, 0, .7854, 1.571, 2.356, 3.142, 3.927, 4.712, 5.498, 6.283

##########sinx###############
sinx = 0, .70711, 1, .70711, 0, -.70711, -1, -.70711, 0, .70711, 1, .70711, 0, -.70711, -1, -.70711, 0

##############cos(x + π/2)#############
cosx = 0, -.70711, -1, -.70711, 0, .70711, 1, .70711, 0, -.70711, -1, -.70711, 0, .70711, 1, .70711, 0 

####plot############

plt.plot(x,sinx,'*--r',label='sin')  
plt.plot(x,cosx,'o--g',label='cos')
plt.show()


```

```python
#subplot

plt.subplot(2,1,1)#two rows, one column,one location 
plt.plot(x,sinx, '*--r', label='sin')
plt.grid() # show grid
plt.legend() #show legend
plt.xlabel('x axis') # x label 
plt.ylabel('y axis') # y label

plt.subplot(2,1,2)# #2 is used for location as it is second plot in the graph
plt.plot(x,cosx, 'o-g', label='cos') 
plt.grid() # show grid
plt.legend() #show legend
plt.xlabel('x axis') # x label 
plt.ylabel('y axis') # y label
############ Legend ####################




plt.show()
```

```python
Mtemp= 22.94, 23.02, 25.68, 19.96, 24.80, 23.98, 22.10, 20.30, 24.20, 22.74, 24.16, 24.94, 22.40, 22.14, 20.84, 25.66, 21.73, 24.49, 24.13, 22.17, 21.73, 20.41, 24.41, 23.95, 20.95, 26.71, 22.81, 23.11, 23.33, 28.83, 23.11, 21.47, 23.97, 24.75, 23.61, 23.08, 21.24, 26.63, 23.88

Days=87, 137, 106, 97, 105, 118, 118, 136, 91, 107, 96, 114, 125, 115, 118, 82, 115, 97, 104, 146, 126, 141, 111, 123, 118, 83, 48, 118, 116, 81, 116, 123, 112, 99, 102, 118, 63, 62, 132

x = np.array(Mtemp)

y=np.array(Days)

colors=np.array(range(1,40,1))

plt.scatter(x,y,c=colors,cmap='viridis',)

plt.colorbar()

plt.show()
```

```python
###############best fit line ##########################

#first method

m, b = np.polyfit(x, y, 1)

print(m)
print(b)
```

```python
plt.plot(x,y,'o')
plt.plot(x,m*x+b)
```
